```r
library(tidyverse) #this is how I load ggplot2, you can also use library(ggplot2)
library(ggthemes)
```

```r
chr_dat = read_csv('chromosome_dat_example.csv') #this is the tidyverse version of read.csv I personally have started using it because I 
                    #find it does a better job of identifying and parsing columns with factors, numbers etc. than the base function
```

```r
head(chr_dat)
```

```r
genome_base = ggplot(NULL) +
    geom_bar(data = chr_dat, 
            aes(x= chr_name, y= chr_len) , 
            stat='identity', 
            fill='grey80', 
            colour='grey80', 
            width=.2) +
    labs(y = "Base pair position", 
            x = "Arctic charr chromosome") +
    theme_minimal() +
    theme_light() +
    theme(panel.border = element_blank(),
            panel.grid.major = element_blank(),
            panel.grid.minor = element_blank())
```

```r
genome_base
```

```r
#this part gives it a minimal theme to get rid of the grey background etc				
chr_only = genome_base + 
    theme(axis.text.x = element_text(angle = 90, hjust = 1),
        axis.ticks.x = element_blank())

chr_only
```

```r
important_segments = read_csv('seg_locations_example.csv')
```

```r
head(important_segments)
```

```r
seg_example = chr_only + 
    geom_segment(data = important_segments, 
        aes(x=chr_name, xend=chr_name, y=mb_start, yend=mb_end), 
        color='blue', 
        size = 2)

seg_example
```

```r
points_example = read_csv('snp_locations_example.csv')
```

```r
head(points_example)
```

```r
marker_locations = chr_only + 
    geom_point(data = points_example, 
        aes(x=chr_name, y=POS), 
        color="grey20",
        shape=22,
        size=.001,
        alpha = 0.8, 
        position=position_jitter(.2))

marker_locations
```

```r
ensemble =  chr_only +
    geom_segment(data = important_segments, 
        aes(x=chr_name, xend=chr_name, y=mb_start, yend=mb_end), 
        color='blue', 
        size = 2) + 
    geom_point(data = points_example, 
        aes(x=chr_name, y=POS), 
        color="grey20",
        shape=22,
        size=.001,
        alpha = 0.8, 
        position=position_jitter(.2))

ensemble
```
